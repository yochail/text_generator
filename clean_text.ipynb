```python
def remove_hashtags(text:str):
  return regex_remove('\#[a-z_]+',text)
```

```python
def remove_mentions(text:str):
  return regex_remove('\@[a-z_\.]+',text)
```

```python
def remove_empty_lines(text:str):
  return regex_remove('\.\r?\n', text)
```

```python
def replace_extra_spaces(text:str):
  return re.sub(' +',' ',text)
def replace_extra_lines(text:str):
  return re.sub('\n+','\n',text)
```

```python
def remove_non_character(text:str):
  return ''.join(c for c in text if c in allowed_chars)
```

```python
def clean_text(text:str):
  text = remove_hashtags(text)
  text = remove_mentions(text)  
  text = remove_empty_lines(text)
  text = replace_extra_spaces(text)
  text = replace_extra_lines(text)
  text = remove_non_character(text)

  return text
```
